# Thai Text Classification - Classical Approach
Uses PyThaiNLP for tokenization and Scikit-Learn tools (TF-IDF, Random Forest) for classification.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# PyThaiNLP for Thai language tokenization
from pythainlp.tokenize import word_tokenize

# --- METHOD A: Classical Approach (TF-IDF + ML) ---
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

# ==========================================
# 1. Load and Preprocess Data
# ==========================================

# NOTE: Replace 'output.csv' with your actual dataset file
# The dataset should have at least two columns: 'text' (the Thai lyrics/messages) and 'label' (the category)

```python
df = pd.read_csv('labeled_songs.csv')
print("Dataset loaded successfully.")

# Extract texts and labels
texts = df['lyric'].astype(str).tolist()
labels = df['label'].tolist()

# Encode labels to integers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(labels)
```

# ==========================================
# 2. METHOD A: Classical Approach
# ==========================================

```python
print("\n--- Method A: Classical Approach (TF-IDF + Random Forest) ---")

# Define a custom tokenizer using PyThaiNLP
def thai_tokenizer(text):
    return word_tokenize(text, engine='newmm')

# Feature Extraction: TF-IDF
vectorizer = TfidfVectorizer(tokenizer=thai_tokenizer, token_pattern=None)
X_tfidf = vectorizer.fit_transform(texts)

# Train-test split
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_tfidf, y_encoded, test_size=0.2, random_state=42)

# Model: Random Forest Classifier (You can also swap with LogisticRegression or MultinomialNB)
classical_model = RandomForestClassifier(n_estimators=100, random_state=42)
classical_model.fit(X_train_c, y_train_c)

# Evaluation
y_pred_c = classical_model.predict(X_test_c)
print("Classical Model Accuracy:", accuracy_score(y_test_c, y_pred_c))
print("\nClassification Report:\n", classification_report(y_test_c, y_pred_c, target_names=label_encoder.classes_))
```
