```python
with open("train.conll","r",encoding="utf-8-sig") as f:
    train= f.read().split("\n\n")
with open("test.conll","r",encoding="utf-8-sig") as f:
    test= f.read().split("\n\n")
with open("validation.conll","r",encoding="utf-8-sig") as f:
    validation= f.read().split("\n\n")
```

```python
tag2idx=[]
def get_idx(t):
    if t not in set(tag2idx):
        tag2idx.append(t)
    return tag2idx.index(t)
```

```python
def make_hf(list_conll2002):
    _words=[]
    _tags=[]
    for i in list_conll2002:
        _w=[]
        _t=[]
        for j in i.splitlines():
            _temp = j.split("\t")
            if len(_temp)!=2:
                continue
            _w.append(_temp[0])
            if _temp[1]=="B-/ORGANIZATION":
                print(i)
            _t.append(get_idx(_temp[1]))#.replace("B-","").replace("I-",""))
        if len(_w)<1:
            continue
        _words.append(_w)
        _tags.append(_t)
    return _words,_tags
```

```python
train_hf = make_hf(train)
val_hf = make_hf(validation)
test_hf = make_hf(test)
```

```python
tag2idx[tag2idx.index("B-T")]="B-TEMPERATURE"
tag2idx[tag2idx.index("I-T")]="I-TEMPERATURE"
```
